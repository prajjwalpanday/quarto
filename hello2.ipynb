```python
#| label: fig-polarw
#| fig-cap: "A line plot on a polar axis"
#| echo: fenced
#| output-location: slide

import numpy as np
import matplotlib.pyplot as plt

r = np.arange(0, 10, 0.01)
theta = 2 * np.pi * r
fig, ax = plt.subplots(
  subplot_kw = {'projection': 'polar'} 
)
ax.plot(theta, r)
ax.set_rticks([0.5, 1, 1.5, 2])
ax.grid(True)
plt.show()
```
